```python
df = df[(df['Number.of.Engines'] == 1) | (df['Number.of.Engines'] == 2)]
```

```python
sns.barplot(data=df2, x='Make', y = 'Total.Uninjured', order=df2.sort_values('Total.Uninjured', ascending=False).Make)
```

```python
df["Total.Fatal.Injuries"].fillna(0, inplace=True)
df["Total.Serious.Injuries"].fillna(0, inplace=True)
df["Total.Minor.Injuries"].fillna(0, inplace=True)
df["Total.Uninjured"].fillna(0, inplace=True)
```

```python
df["total.passengers"] = df["Total.Fatal.Injuries"] + df["Total.Serious.Injuries"] + df["Total.Minor.Injuries"] + df["Total.Uninjured"]
```

```python
df.T
```

```python
df.info()
```

```python
df["fraction.fatal"] = df["Total.Fatal.Injuries"]/df["total.passengers"]
```

```python
df["fraction.uninjured"] = df["Total.Uninjured"]/df["total.passengers"]
```

```python
df3 = df[(df['Make'] == 'Cessna') | (df['Make'] == 'Piper') | (df['Make'] == 'Beech') | (df['Make'] == 'Bell') | (df['Make'] == 'Mooney')]
```

```python
sns.barplot(data=df3, x='Make', y = 'fraction.fatal', order=df3.sort_values('fraction.fatal', ascending=False).Make)
```

```python
sns.barplot(data=df3, x='Make', y = 'fraction.uninjured', order=df3.sort_values('fraction.uninjured', ascending=False).Make)
```
