```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Data/AviationData2.csv')
```

```python
df.T
```

```python
df = df[df['Investigation.Type'] == 'Accident']
```

```python
df['Make'] = df['Make'].str.title()
```

```python
#df = df[(df['Number.of.Engines'] == 3) | (df['Number.of.Engines'] == 4)]
```

```python
df['Survive'] = df['Total.Fatal.Injuries'] == 0
```

```python
df2 = df[(df['Make'] == 'Cessna') | (df['Make'] == 'Piper') | (df['Make'] == 'Beech') | (df['Make'] == 'Bell') | (df['Make'] == 'Mooney') | (df['Make'] == 'Boeing') | (df['Make'] == 'Airbus')]
```

```python
sns.barplot(data=df2, x='Survive', y = 'Make')
```

```python
sns.barplot(data=df2, x='Total.Serious.Injuries', y = 'Make')
```
