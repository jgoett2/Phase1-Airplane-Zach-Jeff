## Data Preparation

```python
## drop Na in make
DF1 = DF.dropna(subset='Make')
```

```python
DF1.info()
```

```python
DF2 = DF1.dropna(subset='Model')
```

```python
DF2.info()
```

```python
DF2['Purpose.of.flight'].value_counts()
```

```python
DF2['Make-Model'] = DF2[['Make', 'Model']].apply(lambda x: ' '.join(x), axis=1)
```

```python
DF2.head(20).T
```

```python
DF3 = DF2[DF2['Make-Model'] == 'Cessna 501' ]
DF3.head(20).T
```

```python
## Make Lookup chart


LookUp = DF2.groupby(['Make-Model'])['Number.of.Engines'].agg(pd.Series.mode)
```

```python
type(LookUp)
```

```python
mv = DF2.groupby('Make-Model')['Number.of.Engines'].transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
```

```python
mv
```

```python
DF2['Number.of.Engines'] = DF2['Number.of.Engines'].fillna(mv)
```
